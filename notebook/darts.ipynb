```python
# 파생변수 제작으로 추가된 변수들이 존재하기에, 다시한번 연속형과 범주형 칼럼을 분리해주겠습니다.
continuous_columns = []
categorical_columns = []

for column in train_for_model.columns:
    if pd.api.types.is_numeric_dtype(train_for_model[column]):
        continuous_columns.append(column)
    else:
        if (column!='구') & (column!='동'): # 구, 동을 제외해준다.
            categorical_columns.append(column)

print("연속형 변수:", continuous_columns)
print("범주형 변수:", categorical_columns)
```

```python
continuous_columns = []
categorical_columns = []

for column in test_for_model.columns:
    if pd.api.types.is_numeric_dtype(test_for_model[column]):
        continuous_columns.append(column)
    else:
        if (column!='구') & (column!='동'): # 구, 동을 제외해준다.
            categorical_columns.append(column)

print("연속형 변수:", continuous_columns)
print("범주형 변수:", categorical_columns)
```

# 인코딩: 레이블

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from tqdm import tqdm


# 각 변수에 대한 LabelEncoder를 저장할 딕셔너리
label_encoders = {}

# Implement Label Encoding
for col in tqdm(categorical_columns):
    lbl = LabelEncoder()

    # 모든 데이터를 문자열로 변환하여 Label-Encoding을 fit
    combined_data = np.unique(np.concatenate((train_for_model[col], test_for_model[col])).astype(str))
    lbl.fit(combined_data)
    
    train_for_model[col] = lbl.transform(train_for_model[col].astype(str))
    test_for_model[col] = lbl.transform(test_for_model[col].astype(str))
    
    label_encoders[col] = lbl  # 나중에 후처리를 위해 레이블인코더를 저장해주겠습니다.

# 결과 확인
print(train_for_model)
print(test_for_model)
```

```python
from sklearn.preprocessing import OneHotEncoder
import pandas as pd

# OneHotEncoder 객체 생성
onehot_encoder = OneHotEncoder(handle_unknown='ignore', sparse_output=False)

# 학습할 데이터의 범주형 변수만 선택
combined_data = pd.concat([train_for_model[categorical_columns], test_for_model[categorical_columns]], axis=0)

# 원핫 인코딩 적용
onehot_encoder.fit(combined_data)

# 훈련 데이터와 테스트 데이터 변환
train_encoded = onehot_encoder.transform(train_for_model[categorical_columns])
test_encoded = onehot_encoder.transform(test_for_model[categorical_columns])

```

```python
# 반환된 배열을 DataFrame으로 변환 (메모리 사용을 감시하면서)
train_encoded_df = pd.DataFrame(train_encoded, columns=onehot_encoder.get_feature_names_out(categorical_columns))
test_encoded_df = pd.DataFrame(test_encoded, columns=onehot_encoder.get_feature_names_out(categorical_columns))
```

```python
import pandas as pd
from darts import TimeSeries
from darts.models import RandomForest
from darts.dataprocessing.transformers import Scaler

# multivariate static covariates (multiple components)
static_covs_columns = ['전용면적(㎡)','층','k-전체동수','k-전체세대수','k-연면적','주차대수',
                        'final_X','final_Y','Nearest_Bus_Station_Distance', 'Nearest_Subway_Station_Distance', 
                    '동수_x', '분양 세대수', '토지 면적(㎡)', 
                    '합계 동수', '세대수', '사용승인일자', 
                    '소형학원수', '중형학원수', '대형학원수', '종합학원수', '입시학원수', '외국어학원수', 
                    '예체능학원수', '독서실수', '정보학원수', '특수교육학원수', '직업기술학원수', '기타학원수','모든학원수',
                    'brand', 'lower_outliers', 'upper_outliers','강남여부', '가격높은동'
]

# '계약년월일' 열을 datetime 형식으로 변환
trainset['계약년월일'] = pd.to_datetime(trainset['계약년월일'])

# 각 일에 대해 유니크한 데이터만 남기기 위해 중복된 '계약년월일' 값을 평균으로 집계
numeric_columns = trainset.select_dtypes(include='number').columns
trainset = trainset.groupby('계약년월일')[numeric_columns].mean().reset_index()

# 타임스탬프 생성: 2007-01-01부터 2023-02-28까지 일별 데이터 생성
date_range = pd.date_range(start='2007-01-01', end='2023-02-28', freq='D')

# '계약년월일'을 인덱스로 설정하고, 타임스탬프를 기준으로 재인덱싱하여 누락된 날짜는 NaN으로 채웁니다.
trainset.set_index('계약년월일', inplace=True)
trainset = trainset.reindex(date_range).reset_index().rename(columns={'index': '계약년월일'})

# 누락된 날짜의 값을 보간 (예: 선형 보간)
trainset['target'] = trainset['target'].interpolate()

# static_covariates 데이터프레임 생성
static_covs_df = trainset[static_covs_columns]

# 타임 시리즈 객체 생성
time_series = TimeSeries.from_dataframe(
    trainset,
    time_col='계약년월일',
    value_cols='target',
    fill_missing_dates=True,
    freq='D',
    fillna_value=None,
    static_covariates=static_covs_df
)

# 데이터 스케일링
scaler = Scaler()
scaled_series = scaler.fit_transform(time_series)

# 훈련 및 검증 데이터 분리
train_series = scaled_series[:'2023-03-01']
val_series = scaled_series['2023-06-30':]

# 글로벌 모델 생성 및 훈련
model = RandomForest(lags=12)  # 또는 최적의 lag 수 선택
model.fit(train_series)

# 예측
forecast = model.predict(n=5, series=train_series)

# 예측 결과 스케일 역변환
final_forecast = scaler.inverse_transform(forecast)

# 결과 출력
print("2023년 3월부터 6월까지 예측 가격:")
print(final_forecast.pd_dataframe())
```
