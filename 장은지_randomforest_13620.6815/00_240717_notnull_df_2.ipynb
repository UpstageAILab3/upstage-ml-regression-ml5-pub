```python
df.loc[df['k-전용면적별세대현황(60㎡이하)'].isna(), 'k-전용면적별세대현황(60㎡이하)'] = 0
df.loc[df['k-전용면적별세대현황(60㎡~85㎡이하)'].isna(), 'k-전용면적별세대현황(60㎡~85㎡이하)'] = 0
df.loc[df['k-85㎡~135㎡이하'].isna(), 'k-85㎡~135㎡이하'] = 0
df.loc[df['k-135㎡초과'].isna(), 'k-135㎡초과'] = 0
df.loc[df['주차대수'].isna(), '주차대수'] = 0
df.loc[df['5층 동수'].isna(), '5층 동수'] = 0
df.loc[df['5층 세대수'].isna(), '5층 세대수'] = 0
df.loc[df['6~10층 동수'].isna(), '6~10층 동수'] = 0
df.loc[df['6~10층 세대수'].isna(), '6~10층 세대수'] = 0
df.loc[df['11~15층 동수'].isna(), '11~15층 동수'] = 0
df.loc[df['11~15층 세대수'].isna(), '11~15층 세대수'] = 0
df.loc[df['16~20층 동수'].isna(), '16~20층 동수'] = 0
df.loc[df['16~20층 세대수'].isna(), '16~20층 세대수'] = 0
df.loc[df['21층이상 동수'].isna(), '21층이상 동수'] = 0
df.loc[df['21층이상 세대수'].isna(), '21층이상 세대수'] = 0
```

```python
# '도로명_fill' 칼럼에서 결측치를 '번지주소_2'의 값으로 채우기
df.loc[df['도로명_fill'].isna(), '도로명_fill'] = df['번지주소_2']
```

```python
df['세대수_sum'] = df['k-전용면적별세대현황(60㎡이하)'] + df['k-전용면적별세대현황(60㎡~85㎡이하)'] + \
                    df['k-85㎡~135㎡이하'] + df['k-135㎡초과']
```

```python
df['세대수_sum_2'] = df['5층 세대수'] + df['6~10층 세대수'] + \
                    df['11~15층 세대수'] + df['16~20층 세대수'] + df['21층이상 세대수']   
```

```python
test = df[['세대수', '번지주소_2', '건물명', '세대수_sum', '세대수_sum_2']].copy()
```

```python
test.loc[(test['세대수'].isna()), '세대수'] = test.loc[(test['세대수'].isna()), '세대수'].fillna(test['세대수_sum_2'])
```

```python
test.loc[(test['세대수'].isna()) & (df['건물명']=='0'), '세대수'] = 10
test.loc[(test['세대수'].isna()) & (df['건물명']=='1'), '세대수'] = 10
```
